```python
from pair_programming import offset_mean

array = [1,2,3,4]
mean = 5 
test_mean = offset_mean(array, mean)

# Making sure the mean is offset by the same amount for each arary value
def test_spacing(array, test_mean):
    for i in range(len(array)-1):
        original_spacing = [array[i+1] - array[i]]
        new_spacing = [test_mean[i+1] - test_mean[i]]
    print(original_spacing==new_spacing)

print(test_spacing(array, test_mean))

# Cheecking that the mean of the output list is equivalent to the offset mean 
def check_mean(test_mean, mean):
    result_mean = sum(test_mean) / len(test_mean)
    print(result_mean == mean)
    
print(check_mean(test_mean, mean))
```
